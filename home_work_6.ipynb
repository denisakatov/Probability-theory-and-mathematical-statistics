Задача 1

Даны значения величины заработной платы заемщиков банка (salary) и значения их поведенческого кредитного скоринга (scoring):

salary = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110]
scoring = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]
Используя только встроенные питоновские функции и структуры данных (т.е. без библиотек numpy, pandas и др.) найдите:

ковариацию (смещённую и несмещённую) этих двух величин,
коэффициент корреляции Пирсона.
Можно затем посчитать те же значения с использованием библиотек, чтобы проверить себя.

```python
# Найдем все  только с использованием встроенных функций python

salary = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110]
scoring = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]

def cov(arr1, arr2):
    ind, result, sum1, sum2 = 0, 0, 0, 0
    for i in arr1:
        sum1 += i
        mean1 = sum1/len(arr1)
    for m in arr2:
        sum2 += i
        mean2 = sum2/len(arr2)
    while ind != len(arr1):
        m = (arr1[ind]-mean1)*(arr2[ind]-mean2)
        result += m
        ind += 1
    return result/(len(arr1)-1)

print(cov(salary, scoring))
def std(arr):
    sum, result = 0, 0
    for i in arr:
        sum += i
        mean = sum/(len(arr))
    for i in arr:
        disp = ((i-mean)**2)
        result += disp/(len(arr)-1)
    return result**0.5

    
def pearson(cov):
    return cov/(std(salary)*std(scoring))

print(pearson(cov(salary, scoring)))
```

```python
# С частичным использованием numpy
import numpy as np
import pandas as pd


print('Принты только для праверки правильности функции стандартного отклонения')
print(std(salary), np.std(salary, ddof = 1))
print(std(scoring), np.std(scoring, ddof = 1))



```

```python
# С частичным использованием numpy

print(np.std(scoring, ddof = 1))
n = len(salary)
sigma_salary_scoring = ((salary - np.mean(salary))*(scoring-np.mean(scoring))).sum()/(n-1) # Ковариация
cov = sigma_salary_scoring/(np.std(salary, ddof = 1)*np.std(scoring, ddof = 1)) # коэффициент корреляции Пирсона.
sigma_salary_scoring, cov
```

```python
np.cov(salary, scoring, ddof = 0) # смещенная
```

```python
np.cov(salary, scoring, ddof = 1) # несмещенная. Проверяем - совпадает
```

### Задача 2

Проведите тест на значимость коэффициента корреляции Пирсона, найденного в предыдущей задаче. Что для этого нужно знать:

Нулевая гипотеза: реальный коэффициент корреляции равен 0. Альтернативная гипотеза двухсторонняя.
Статистика: t = r * sqrt(n - 2) / sqrt(1 - r ** 2), где r - коэффициент корреляции Пирсона, посчитанный по выборке.
В предположении верности нулевой гипотезы эта статистика имеет распределение Стьюдента с параметром df = n - 2.

```python
from scipy import stats
alpha = 0.05
tl = stats.t.ppf(alpha/2, df = n - 2)
tr = stats.t.ppf(1-alpha/2, df = n - 2)
tl, tr
```

```python
r = cov
t0 = r*np.sqrt(n-2)/np.sqrt(1-r**2)
t0
```

```python
vr = 1-stats.t.cdf(t0, df = n-2)
vl = stats.t.cdf(t0, df = n-2)

print(2*min(vr, vl))
```

Вывод: Нулевая гипотеза отвергается, так как результат находится в критической области и значение p-value < 0.05

```python
import numpy as np
from scipy import stats
```

```python
iq = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])
n = len(iq)
```

```python
alpha = 0.05
```

```python
tl = stats.t.ppf(alpha/2, df = n-1)
tr = stats.t.ppf(1-alpha/2, df = n-1)
```

```python
tl, tr
```

```python
f'Доверительный интервал: {np.mean(iq) + tl*np.std(iq)/np.sqrt(n), np.mean(iq) + tr*np.std(iq)/np.sqrt(n)}'
```
