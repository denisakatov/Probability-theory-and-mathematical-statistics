## Задача 1

Контрольная работа состоит из пяти вопросов. На каждый вопрос приведено четыре варианта ответа, один из которых правильный. Составьте закон распределения числа правильных ответов при простом угадывании. Найдите M(X), D(X).

```python
import numpy as np
from math import factorial as fc
```

```python
questions = np.arange(0,6) #Количество возможных вариантов (ничего не ответил, ответил правильно на все 5 вопросов)
n = 5
answers = 4
```

```python
p = 1/answers # вероятность ответить правильно на один вопрос
q = 1-p # Вероятность ответить не правильно
print(p, q)
```

```python
questions
```

```python
summa = 0
for k in questions:
    result = fc(n)/(fc(k)*fc(n-k))*p**(k)*q**(n-k)
    print(f'Вероятность того, что правильно ответит на {k} вопрос(а) из {n} вопросов (P(x={k})) - {result}')
    summa += result
print(summa)
```

```python
m = n*p # Математическое ожидание
m
```

```python
d = n*p*q
d
```

## Задача 2

Пользователь получает в среднем 10 писем со спамом на свой почтовый ящик за сутки. Найти число N, такое, что с вероятностью 0.95 пользователь получит не более N писем со спамом за текущий день.

```python
from math import e
```

```python
def poisson(N):
    p = ((10**N)*(e**(-10)))/(fc(N)) # Формула Пуассона
    return p
poisson(12)
```

```python
counts = [i for i in range(0, 30)]
print(counts)
```

```python
a = 0
for k in counts:
    a += poisson(k)
    print(k+1, 0.95-a)
```

## Ответ: На 16 день пользователь получит более 0,95 (15 дней он не получит писем с вероятностью больше 0,95)

## Задача 3

Производятся выстрелы по мишени. Вероятность попадания в мишень при одном выстреле равна 0.01. Сколько выстрелов нужно сделать чтобы быть уверенным с вероятностью 0.9, что хотя бы 6 раз будет совершено попадание?

Подсказка. 1) "Вероятность попасть k раз при n выстрелах" - на какое распределение это похоже? 2) А если нам нужна вероятность P(X >= k), а не P(X = k)? 3) Здесь предстоит немножко покодить.

```python
p = 0.01 # вероятность попадания с одного раза
q = 1-p
k = 6 # количество попаданий
n = 6 # искомое количество выстрелов (менее 6 быть не может)
t = 0
def search(n):
    P = 0
    result = 0
    k = 6
    p = 0.01
    while k <= n: # пока n не равно k
        P = ((fc(n))/(fc(k)*fc(n-k))) * (p**k) * ((1-p)**(n-k)) # Формула Бернулли
        result += P # складываем результат от n до k
        k += 1 
    return result
```

```python
while t < 0.9:
    n += 1
    t = search(n)
    print(n, t)
```
